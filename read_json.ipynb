```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_json('data/atlas.json')
df
```

```python
df = df[['id','name','path','center']]
df
```

```python
# e.g. corners = [(2.0, 1.0), (4.0, 5.0), (7.0, 8.0)]
def calc_area(v):    
    #|print(v)
    a = list(v.keys())
    a = a[0]
    corners = v[a]
    n = len(corners) # of corners
    area = 0.0
    for i in range(n):
        j = (i + 1) % n
        area += corners[i][0] * corners[j][1]
        area -= corners[j][0] * corners[i][1]
    area = abs(area) / 2.0
    return area
```

```python
df['area'] = df['path'].apply(lambda x: calc_area(x))
df.head()
```

```python
df.area.sort_values(ascending=False,inplace= False)
```
